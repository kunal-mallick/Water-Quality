# Feature Engineering & Feature Extraction

## Attaching Resources

### Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

### Adding Dataset

```python
wp = pd.read_csv('https://raw.githubusercontent.com/kunal-mallick/Kaggle-Project/main/Water%20Quality(Drinking%20Water%20Potability)/src/main/com/Data%20Preprocessing%20%26%20EDA/EDA.csv')
#wp.drop(columns='	Unnamed: 0', inplace = True)

wp
```

## Splitting The Data Into Target and Independent set

```python
x = wp.iloc[:,:-1]
y = wp['Potability']
```

```python
x
```

### Normalizing The Dataset

```python
sc = StandardScaler()
scaled_x  = pd.DataFrame(sc.fit_transform(x), columns=x.columns)
```

```python
scaled_x
```

### Splitting The Data Into Training And Testing Set

```python
xtrain,xtest,ytrain,ytest = train_test_split(scaled_x,y,train_size=0.75,random_state=1)
```

```python
xtrain
```

## Exporting Dataframe

```python
xtrain.to_csv('xtrain.csv', index=False)
xtest.to_csv('xtest.csv', index=False)
ytrain.to_csv('ytrain.csv', index=False)
ytest.to_csv('ytest.csv', index=False)
```
